Create three variables: `num1`, `num2`, and `operator` (as a string: `"+"`, `"-"`, `"*"`, or `"/"`). Using if-elif-else, perform the correct calculation based on the operator and print the result.

````python
# Write your code here
num1 = 10
num2 = 5
operator = "/"

if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    print(num1 / num2)
else:
    print("Invalid operator")
````

### Exercise B2: Leap Year Checker

Create a variable `year`. Write an if-else statement to check if it's a **leap year**.

Rule: A year is a leap year if it's divisible by 4, EXCEPT years divisible by 100 — unless they're also divisible by 400.

💡 *Hint: You'll need to combine `and`, `or`, and `!=` together.*

````python
# Write your code here
year = 2024

if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
````

<summary>💡 Click to see Solution</summary>


### Exercise B3: FizzBuzz (Single Number)

Create a variable `num`. Using if-elif-else:
- If divisible by both 3 and 5 → print "FizzBuzz"

- Otherwise → print the number itself

````python
# Write your code here
num = 17

if num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")
elif num % 3 == 0:
    print("Fizz")
elif num % 5 == 0:
    print("Buzz")
else:
    print(num)
````
