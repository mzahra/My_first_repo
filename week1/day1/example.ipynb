```python
# Defining a class
class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course
        self.grades = []
    
    def add_grade(self, grade):
        """Add a grade to the student's record"""
        self.grades.append(grade)
    
    def get_average(self):
        """Calculate and return the average grade"""
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)
    
    def introduce(self):
        """Return a formatted introduction string"""
        return f"Hi, I'm {self.name}, {self.age} years old, studying {self.course}"
 
# Creating and using objects
student1 = Student("Alice", 25, "AI Bootcamp")
student1.add_grade(85)
student1.add_grade(90)
student1.add_grade(88)
 
print(student1.introduce())  # "Hi, I'm Alice, 25 years old, studying AI Bootcamp"
print(f"Average: {student1.get_average()}")  # 87.67
```
